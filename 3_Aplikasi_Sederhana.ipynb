#Aplikasi Sederhana Menghitung Volume Bola

```python
jari_jari = int(input("Masukkan jari-jari bola (cm): "))
pi = 22/7
volume_bola = 4/3 * pi * jari_jari ** 3
print(f"Volume bola adalah {volume_bola} cm^3. ")
```

#Menghitung Volume Kerucut

```python
jari_jari = int(input("Masukkan jari_jari kerucut (cm): "))
tinggi = int(input("Masukkan tinggi kerucut (cm): "))
pi = 22/7
volume_kerucut = 1/3 * pi * jari_jari ** 2 * tinggi
print(f"Volume kerucut adalah {volume_kerucut} cm^2. "))
```

```python
jari_jari = int(input("Masukkan jari-jari dari kerucut (dm): "))
tinggi = int(input("Masukkan tinggi dari kerucut (dm): "))
pi = 3.14
volume_kerucut = 1/3 * pi * jari_jari ** 2 * tinggi
print(f"volume kerucut adalah {volume_kerucut} dm^2. ")
```
