#Fungsi Parameter

```python
nama_siswa_1 = "Daffanes"
nama_siswa_2 = "Adzka"
nama_siswa_3 = "Haidar"

print(nama_siswa_1, nama_siswa_2, sep="-")
```

```python
# Latihan
# Silakan tampilkan keluaran/output berupa:
# Haidar*Adzka*Daffanes*Adzka
```

```python
print(nama_siswa_3, nama_siswa_2, nama_siswa_1, nama_siswa_2, sep="*")
```

Latihan Soal :
Gunakan Parameter end untuk menampilkan tulisan sebagai berikut :
Haidar#Adzka#Daffanes#Adzka

```python
print(nama_siswa_3, end="#")
print(nama_siswa_2, end="#")
print(nama_siswa_1, end="#")
print(nama_siswa_2)
```
